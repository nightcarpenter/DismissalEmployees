### Функция для вывода распределения категориального признака

```python
def show_cat_feature(df, feature, 
                     feature_name='', 
                     cut_count_to_barplot=5):
    '''
        Строит круговую диаграмму или барплот для категориального признака.
        Показывает статистические параметры.
        
        Args:
            df (pandas.DataFrame) - датафрейм, в котором находится признак
            feature (str) - название признака
            feature_name (str, optional) - имя признака для подписи графика default feature_name=feature
            cut_count_to_barplot (int, optional) - минимальное количество категорий для построения барплота 
                default cut_count_to_barplot=5
    '''
    
    # Ставим дефолтное название признака для подписи графика
    if feature_name == '':
        feature_name = feature
    
    # Создадим сводную таблцу для построения гистограммы или барплота
    pivot = df.pivot_table(index=feature, values='id', aggfunc='count')
    pivot.columns = ['customers']
    pivot = pivot.sort_values(by='customers')
    
    # Строим круговую диаграмму, если категорий меньше 5
    if len(pivot) < 5:
        # Создаём фигуру и ось
        fig, ax = plt.subplots()
        # Подписываем диаграмму
        ax.set_title(feature_name, fontsize=11)
        # Отрисовываем диаграмму
        ax.pie(pivot['customers'], 
               labels=pivot.index, 
               autopct='%1.2f%%',
               explode=([0.03] * len(pivot)))
    
    # Строим барплот, если категорий больше или равно 5
    else:
        # Создаём фигуру и оси
        fig, ax = plt.subplots(figsize=(6, 3))
        # Устанавливаем размер шрифта
        plt.rcParams.update({'font.size': 9})
        # Наносим сетку, подписываем оси
        ax.grid(axis='x')
        ax.set(title=feature_name,
               xlabel='количество сотрудников')
        # Отрисосываем графики
        ax.barh(pivot.index, pivot['customers'], linewidth=1.0, edgecolor='k', height=0.7)
    
    # Выводим график на экран
    plt.show()
```

<a id="Функция_гистограммы_распределения_количественного_признака"></a>
### Функция для вывода гистограммы распределения количественного признака

```python
def show_num_feature(df, feature, 
                     feature_name='', 
                     bins=100,
                     boxplot_flag=True,
                     describe_flag=True):
    '''
        Строит гистограмму распределения и ящик с усами для количественного признака.
        Показывает статистические параметры.
        
        Args:
            df (pandas.DataFrame) - датафрейм, в котором находится признак
            feature (str) - название признака
            feature_name (str, optional) - имя признака для подписи графика default feature_name=feature
            bins (int, optional) - количество корзин на гистограмме default bins=100
            boxplot_flag (bool, optional) - показывать ящик с усами default boxplot_flag=True
            describe_flag (bool, optional) - показывать ящик с усами default describe_flag=True
    '''
    
    # Устанавливаем размер шрифта
    plt.rcParams.update({'font.size': 9})
    
    # Ставим дефолтное название признака для подписи графика
    if feature_name == '':
        feature_name = feature
    
    # Задаём фигуру и оси
    if boxplot_flag:
        fig, axs = plt.subplots(2, 1, figsize=(8, 5))
        ax_1 = axs[0]
        ax_2 = axs[1]
    else:
        fig, ax_1 = plt.subplots(1, 1, figsize=(8, 3))
    
    # Отрисовываем гистограмму
    ax_1.set_title(f'Гистограмма распределения для признака: {feature_name}', fontsize=11)
    ax_1.set_xlabel(f'{feature_name}')
    ax_1.set_ylabel('количество сотрудников', fontsize=9)
    ax_1.grid()
    ax_1.hist(df[feature], bins=bins, linewidth=1.0, edgecolor='k')
    
    # Отрисовываем ящик с усами, если стоит флаг 
    if boxplot_flag:
        ax_2.set_title(f'Ящик с усами для признака: {feature_name}', fontsize=9)
        ax_2.set_xlabel(f'{feature_name}', fontsize=9)
        ax_2.grid()
        ax_2.boxplot(df[feature], vert=False)
    
    # Добавляем строчку, что бы не перекрывались подписи осей
    plt.tight_layout(pad=2.0)
        
    # Выводим график на экран
    plt.show()
    
    # Смотрим статистические параметры
    if boxplot_flag:
        display(pd.DataFrame(df[feature].describe()))
```

<a id="Функция_анализа_портрета"></a>
### Функция для анализа портрета уволившегося сотрудника

```python
def show_pivot(df, feature, title):
    '''
    Строит сводную таблицу и график barplot,
    для демонстрации зависимости процента увольнений от входного признака.

    Args:
        df (pandas.DataFrame) - датафрейм, в котором находится признак
        feature (str) - название признака
        title (str) - название итогового графика
    '''
    
    pivot = df.pivot_table(index=feature, columns='quit', values='id', aggfunc='count')
    pivot.index = pivot.index.astype('str')
    pivot.columns.name = ''
    pivot['all'] = pivot['yes'] + pivot['no']
    pivot['ratio'] = round(pivot['yes'] / pivot['all'] * 100, 2)
    pivot = pivot.sort_values(by='ratio', ascending=False)
    
    plt.rcParams['figure.figsize'] = (6, 3)
    plt.rcParams.update({'font.size': 9})

    # Построение графика без параметра hue
    barplot = sns.barplot(data=pivot.reset_index(), x='ratio', y=feature, palette='viridis')
    plt.title(title)
    plt.xlabel('Процент уволившихся')
    plt.ylabel('')
    
    # Создаем кастомные цвета для легенды, соответствующие цветам столбцов
    patches = barplot.patches
    handles = [plt.Rectangle((0, 0), 1, 1, color=patch.get_facecolor()) for patch in patches]
    labels = [f'{ratio}%' for ratio in pivot['ratio']]
    
    # Настраиваем легенду
    plt.legend(handles=handles[:len(labels)], labels=labels, title='ratio', bbox_to_anchor=(1, 1), loc='upper left')
    
    plt.show()
    
    display(pivot)
```

### Класс для создания семплера-пустышки

```python
'''
    Создаём класс-пустышку для семплера.
    Нужен для того, что бы сэмплер не выдавал ошибке при попытке вызвать fit и predict,
    при значении paththrough.
'''
class Passthrough(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return X
```

```python
'''
    Нуден для того что бы можно было кодировать категориальные признаки
    по технологии embedding, с которой хорошо работают модели градиентного бустинга
'''
class EmbeddingEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, embedding_dim=2, handle_unknown='ignore', random_state=None):
        self.embedding_dim = embedding_dim
        self.handle_unknown = handle_unknown
        self.random_state = random_state
        # Словарь уникальных значений категорий
        self.embeddings_ = {}
        # Словарь моделей встраиваний
        self.models_ = {}
        # Подавляем вывод журнала
        tf.get_logger().setLevel('ERROR')
        
        if random_state is not None:
            np.random.seed(random_state)
            tf.random.set_seed(random_state)

    def fit(self, X, y=None):
        # Обучение моделей встраиваний для каждого столбца
        original_log_level = tf.get_logger().level
        
        # Проходим по каждому столбцу данных
        for i, col in enumerate(X.T):
            # Находим и сохраняем уникальные значения в столбце
            unique_vals = np.unique(col)
            self.embeddings_[i] = unique_vals
            # Входной слой для модели
            input_layer = Input(shape=(1,))
            # Слой встраивания
            embedding_layer = Embedding(input_dim=len(unique_vals) + 1, 
                                        output_dim=self.embedding_dim, 
                                        input_length=1)(input_layer)
            # Развертывание вектора встраивания в плоский массив
            flatten_layer = Flatten()(embedding_layer)
            # Создаем, компилируем и сохраняем модель для текущего столбца
            model = Model(inputs=input_layer, outputs=flatten_layer)
            model.compile(optimizer='rmsprop', loss='mse')
            self.models_[i] = model
        # Восстанавливаем оригинальный уровень логирования TensorFlow
        tf.get_logger().setLevel(original_log_level) 
        return self
    
    def transform(self, X):
        # Преобразование данных с использованием обученных моделей встраиваний.
        original_log_level = tf.get_logger().level
        
        # Список для хранения преобразованных данных
        X_transformed = []
        # Проходим по каждому столбцу данных
        for i, col in enumerate(X.T):
            # Получаем модель для текущего столбца
            model = self.models_[i]
            
            # Список для хранения индексов категорий
            col_indices = []
            # Преобразуем каждое значение в столбце в индекс
            for val in col:
                if val in self.embeddings_[i]:
                    col_indices.append(np.where(self.embeddings_[i] == val)[0][0])
                else:
                    # Не забываем про индекс для неизвестной категории
                    if self.handle_unknown == 'ignore':
                        col_indices.append(len(self.embeddings_[i]))
                    else:
                        raise ValueError(f"Unknown category '{val}' in column {i}.")
           
            # Преобразуем список индексов в массив
            col_indices = np.array(col_indices)
            # Получаем встраивания для индексов
            col_embedded = model.predict(col_indices, verbose=0)
            # Добавляем встраивания в список
            X_transformed.append(col_embedded)
        
        # Возвращаем преобразованные данные как один массив
        return np.hstack(X_transformed)
    
    
    def get_feature_names_out(self, input_features=None):
        # Формирование списка имен новых признаков после встраивания
        feature_names = []
        for i in range(len(self.embeddings_)):
            for dim in range(self.embedding_dim):
                feature_names.append(f"col_{i}_dim_{dim}")
        return np.array(feature_names)
```

<a id="Класс_для_создания_метрики"></a>
### Класс для создания метрики

```python
class MetricSmape:
    def __init__(self):
        self.smape = make_scorer(self.make_smape, 
                                 greater_is_better=self.is_max_optimal())
    
    
    # Метод для создания метрики и последующего её использования для рассчётов
    def make_smape(self, y_true, y_pred, weight=None):
        if len(y_pred.shape) == 1:
            y_pred = self.remake_pred_array(y_pred)
        
        smape = self.custom_logloss_metric(y_true, y_pred)
        
        return np.mean(smape)
    
    
    # Метод нужен потому что разные модели отправляют данные в разных форматах
    def remake_pred_array(self, pred_array):
        new_pred_array = []
        for y in pred_array:
            new_pred_array.append([y])
        new_pred_array = np.asarray(new_pred_array)
        
        return new_pred_array
    
    
    # Метод, чтобы использовать в качестве eval_metric в catbost
    def evaluate(self, approxes, target, weight):
        approx = approxes[0]
        assert len(target) == len(approx)

        if weight is None:
            weight = np.ones(len(target))
        
        smape = self.custom_logloss_metric(approx, target)
        
        # Возвращаем кортеж (сумма ошибок, сумма весов)
        weighted_smape = np.sum(weight * smape) / np.sum(weight)
        
        return np.sum(weight * smape), np.sum(weight)
    
    
    # Метод для подсчёта метрики, используется для catboost как loss_function
    def custom_logloss_metric(self, y_true, y_pred):
        error = np.abs(y_true - y_pred)
        scale = (np.abs(y_true) + np.abs(y_pred)) / 2
        smape = error / scale * 100
        return smape
    
    
    # Метод для подсчёта производных, используется для catboost как loss_function
    def calc_ders_range(self, approxes, targets, weights):
        ders = []
        for approx, target in zip(approxes, targets):
            scale = (np.abs(target) + np.abs(approx)) / 2
            
            # Первая производная
            numerator = 2 * np.sign(approx - target)
            denominator = scale
            der1 = numerator / denominator
            
            # Вторая производная
            numerator = 2 * (np.abs(target) + np.abs(approx)) - 4 * np.sign(approx - target)
            denominator = scale ** 2
            der2 = numerator / denominator
            
            ders.append((der1, der2))
        return ders
    
    
    # Отдельный метод нужен использования класса в качестве eval_metric в catbost
    def is_max_optimal(self):
        return False
    
    
    # Метод нужен только для использования класса в качестве eval_metric в catbost
    def get_final_error(self, error, weight):
        return error / (weight + 1e-38)
```

### Класс для feature engineering

```python
# Класс для изменения и удаления признаков, а также создания новых признаков
class FeatureEngineering:
    def __init__(self):
        pass
    
    
    # Основной метод, проводящий все преобразования, в зависимости от флагов
    def run(self, df, 
            drop_level_flag=False, 
            drop_workload_flag=False,): 
        
        self.df = df.copy()
        self.missing_values_preprocessor()
        self.drop_level(drop_level_flag)
        self.drop_workload(drop_workload_flag)
        
        return self.df
    
    
    # Метод для приведения пропусков к одному виду
    def missing_values_preprocessor(self):
        object_columns_list = self.df.select_dtypes(include='object').columns.tolist()
        for column_name in object_columns_list:
            self.df[column_name] = self.df[column_name].str.strip().replace('', np.nan)
    
    
    # Метод для удаления признака revenues_current_month (из-за высокой корреляции)
    def drop_level(self, drop_level_flag):
        if drop_level_flag:
            self.df = self.df.drop(columns='level')
    
    
    # Метод для удаления признака revenues_current_month (из-за высокой корреляции)
    def drop_workload(self, drop_workload_flag):
        if drop_workload_flag:
            self.df = self.df.drop(columns='workload')
```

<a id="Класс_для_пайплайна"></a>
### Общий класс для создания пайплайна

```python
'''
    Создаём класс для использования библиотеки optuna в котором прописываем пайплайн, 
    вместе с переменными параметрами, которые будем подбирать.
    Это базовый кламм, от него будут наследоваться классы для регрессии и клсссификации
'''
class Objective:
    def __init__(self, X_train, y_train, cv_num):
        self._X_train = X_train
        self._y_train = y_train
        self.cv_num = cv_num
        
        self.model_name = None
        self.best_pipeline = None
        self.score = None
        self.n_catboost_iters = None
        
        self.cat_features = X_train.select_dtypes(include=['object']).columns.tolist()
        self.feature_engineering = FeatureEngineering()
        self.tree_models_list = ['DecisionTreeRegressor', 'DecisionTreeClassifier', 
                                 'RandomForestRegressor', 'RandomForestClassifier']
    
    
    # Основная функция, вызывается при запуске подбора параметров методом study.optimize
    def __call__(self, trial):
        self.create_feature_engineering_params(trial)
        
        # Создаём копии массива и словаря категорий для одного триала
        self.cat_features_trial = self.cat_features.copy()
        self.category_order_trial = self.category_order.copy()
        # Удаляем из словаря категорий для OrdinalEncoder признаки, если они не участвуют в триале
        if (self._feature_engineering_param['drop_workload_flag'] & 
            ('workload' in self.cat_features_trial) & ('workload' in self.category_order_trial)):
            self.cat_features_trial.remove('workload')
            self.category_order_trial.pop('workload')
        if (self._feature_engineering_param['drop_level_flag'] & 
            ('level' in self.cat_features_trial) & ('level' in self.category_order_trial)):
            self.cat_features_trial.remove('level')
            self.category_order_trial.pop('level')
        
        self.create_preprocessor_params(trial)
        self.create_sampler_params(trial)
        self.create_model_with_model_params(trial)
        self.create_pipeline(trial)
        
        # Для классификации стратифицируем кроссвалидацию для регрессии нет
        if isinstance(self.kf, KFold):
            split_args = (self._X_train,)
        elif isinstance(self.kf, StratifiedKFold):
            split_args = (self._X_train, self._y_train)
        
        # Запускаем кроссвалидацию вручную, т.к. кроссвалидация catboost не поддерживает работу с пайплайном
        self.n_catboost_iters = None
        met_cv_list = []
        n_iterations_list = []
        for i, (train_index, val_index) in enumerate(self.kf.split(*split_args)):
            X_train, X_val = self._X_train.iloc[train_index], self._X_train.iloc[val_index]
            y_train, y_val = self._y_train.iloc[train_index], self._y_train.iloc[val_index]
            X_train.reset_index(drop=True, inplace=True)
            y_train.reset_index(drop=True, inplace=True)
            
            # Для catboost отдельно готовим пулы и обучаем модель
            if self.model_name in ['CatBoostRegressor', 'CatBoostClassifier']:
                # Сначала тдельно преобразовываем данные, потом обучаем модель на преобразованных пулах
                X_train_pre = X_train.copy()
                X_val_pre = X_val.copy()
                y_train_pre = y_train.copy()
                y_val_pre = y_val.copy()
                
                # Подготовить днные одним действием не получится из-за использования ImbPipeline 
                X_train_pre = self._pipe_final.named_steps['feature_engineering'].fit_transform(X_train_pre)
                X_val_pre = self._pipe_final.named_steps['feature_engineering'].transform(X_val_pre)
                X_train_pre = self._pipe_final.named_steps['preprocessor'].fit_transform(X_train_pre, 
                                                                                         y_train_pre)
                X_val_pre = self._pipe_final.named_steps['preprocessor'].transform(X_val_pre)
                
                # Если задан семплер и кодировщик категориальных признаков, то семплируем
                if (not isinstance(self._sampler, Passthrough)) & (not isinstance(self._cat_encoding, Passthrough)):
                    X_train_pre, y_train_pre = self._pipe_final.named_steps['sampler'].fit_resample(X_train_pre, y_train_pre)
                
                # Создаём пулы из преобразованных данных
                train_pool = Pool(data=X_train_pre, label=y_train_pre)
                val_pool = Pool(data=X_val_pre, label=y_val_pre)
                
                # Отдельно обучаем модель на уже преобразованных данных
                self._pipe_final.named_steps['model'].fit(train_pool, 
                                                          eval_set=val_pool, 
                                                          use_best_model=True,
                                                          early_stopping_rounds=100)
                 # Получаем количество итераций
                n_iterations_list.append(self._pipe_final.named_steps['model'].get_best_iteration())
                
            # Просто обучаем пайплайни, если модель не catboost
            else:
                self._pipe_final.fit(X_train, y_train)
            

            
            # Получаем предсказания для одной кроссвалидационной итерации и помещаем их в список
            met = self.score(self._pipe_final, X=X_val, y_true=y_val)
            met_cv_list.append(met)
        
        # Сохраняем среднее количество итераций в одном фолде для CatBoost
        if self.model_name in ['CatBoostRegressor', 'CatBoostClassifier']:
            self.n_catboost_iters = round(np.mean(n_iterations_list))
            trial.set_user_attr('iterations', self.n_catboost_iters)
        
        # Возвращаем среднюю метрику по всем фолдам кроссвалидации
        score_value = np.array(met_cv_list).mean()
        return score_value
    
    
    # Задаём тренировочную выборку
    def set_train_and_test_data(self, X_train, y_train):
        self._X_train = X_train
        self._y_train = y_train

    
    # Метод для изменения и удаления признаков, а также создания новых признаков
    def do_feature_engineering(self, df, drop_workload_flag, drop_level_flag):
        df = self.feature_engineering.run(df, drop_workload_flag, drop_level_flag)
        return df
    
    
    # Метод для создания флагов для feature_engineering (для подбора параметров)
    def create_feature_engineering_params(self, trial):
        self._feature_engineering_param ={
            'drop_workload_flag': trial.suggest_categorical('drop_workload_flag', [True, False]),
            'drop_level_flag': trial.suggest_categorical('drop_level_flag', [True, False]),
        }

    
    # Метод для выбора способа кодирования и масштабирования (для подбора параметров)
    def create_preprocessor_params(self, trial):
        self._num_encoding_param = {
            'num_encoding': trial.suggest_categorical('num_encoding', ['MinMaxScaler', 
                                                                       'StandardScaler', 
                                                                       'RobustScaler', 
                                                                       'passthrough']),
        }
        
        # Для CatBoost устанвавливаем три способа кодирования категориальных признаков
        if (self.model_name in ['CatBoostRegressor', 'CatBoostClassifier']):
            self._cat_encoding_param = {
                'cat_encoding': trial.suggest_categorical('cat_encoding', ['OrdinalEncoder', 
                                                                           'OneHotEncoder', 
                                                                           'TargetEncoder',
                                                                           'EmbeddingEncoder'])
            }
        
        # Для RandomForest и DecisionTree устанвавливаем три способа кодирования категориальных признаков
        elif (self.model_name in self.tree_models_list):
            self._cat_encoding_param = {
                'cat_encoding': trial.suggest_categorical('cat_encoding', ['OrdinalEncoder', 
                                                                           'OneHotEncoder', 
                                                                           'TargetEncoder'])
            }
            
        
        # Для остальных моделей устанвавливаем один способ кодирования категориальных признаков
        else:
            self._cat_encoding_param = {
                'cat_encoding': trial.suggest_categorical('cat_encoding', ['OrdinalEncoder', 
                                                                           'OneHotEncoder'])
            }

        if self._cat_encoding_param['cat_encoding'] == 'OrdinalEncoder':
            categories = [self.category_order_trial[col] for col in self.cat_features_trial]
            self._cat_encoding = OrdinalEncoder(categories=categories, 
                                                handle_unknown='use_encoded_value', 
                                                unknown_value=-1)
        elif self._cat_encoding_param['cat_encoding'] == 'OneHotEncoder':
            self._cat_encoding = OneHotEncoder(drop='first', handle_unknown='ignore')
        elif self._cat_encoding_param['cat_encoding'] == 'TargetEncoder':
            self._cat_encoding = TargetEncoder(handle_unknown='value')
        elif self._cat_encoding_param['cat_encoding'] == 'EmbeddingEncoder':
            self._embedding_dim = trial.suggest_int('embedding_dim', 5, 40)
            self._cat_encoding = EmbeddingEncoder(embedding_dim=self._embedding_dim, 
                                                  handle_unknown='ignore', 
                                                  random_state=RANDOM_STATE)
        else:  
            self._cat_encoding = Passthrough()

        if self._num_encoding_param['num_encoding'] == 'MinMaxScaler':
            self._num_encoding = MinMaxScaler()
        elif self._num_encoding_param['num_encoding'] == 'StsndardScaler':
            self._num_encoding = StandardScaler()
        elif self._num_encoding_param['num_encoding'] == 'RobustScaler':
            self._num_encoding = RobustScaler()
        else:
            self._num_encoding = Passthrough()

    
    # Метод для создания пайплайна
    def create_pipeline(self, trial):
        self._cat_pipe = ImbPipeline(
            [
                ('simple_imputer_before_cat', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                ('cat', self._cat_encoding),
                ('simple_imputer_after_cat', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
            ]
        )
        
        self._num_pipe = ImbPipeline(
            [
                ('simple_imputer_before_num', SimpleImputer(missing_values=np.nan, strategy='median')),
                ('num', self._num_encoding)
            ]
        )

        self._preprocessor = ColumnTransformer(
            [
                ('cat', self._cat_pipe, make_column_selector(dtype_include=object)),
                ('num', self._num_pipe, make_column_selector(dtype_include=np.number))
            ], 
            remainder=Passthrough()
        )
        
        self._pipe_final = ImbPipeline(
            [
                ('feature_engineering', FunctionTransformer(self.do_feature_engineering, 
                                                            kw_args={**self._feature_engineering_param})),
                ('preprocessor', self._preprocessor),
                ('sampler', self._sampler),
                ('model', self._model)
            ]
        )
    
    
    # Метод для создания копии пайплайна с новой необученной моделью и измененными параметрами
    def change_model_params(self, pipeline, new_params):
        # Создаем копию пайплайна без копирования самой модели
        new_pipeline = ImbPipeline([
            (name, step) for name, step in pipeline.steps if name != 'model'
        ])

        # Извлекаем текущие параметры модели
        current_model = pipeline.named_steps['model']
        current_params = current_model.get_params()

        # Обновляем параметры новыми значениями
        current_params.update(new_params)

        # Создаем новый экземпляр модели с обновленными параметрами
        if self.model_name == 'CatBoostRegressor':
            new_model = CatBoostRegressor(**current_params)
        elif self.model_name == 'CatBoostClassifier':
            new_model = CatBoostClassifier(**current_params)

        # Добавляем новую модель в пайплайн
        new_pipeline.steps.append(('model', new_model))

        return new_pipeline
    
    
    # Метод автоматически вызывается после каждого триала, 
    # нужна для сохранения пайплайна с лучшей комбинацией параметров
    def callback(self, study, trial):
        if study.best_trial.number == trial.number:
            self.best_pipeline = self._pipe_final
            
            # Заменяем количество итераций на среднее по кроссвалидации для Catboost
            if self.model_name in ['CatBoostRegressor', 'CatBoostClassifier']:
                new_params = {'iterations': self.n_catboost_iters}
                self.best_pipeline = self.change_model_params(self.best_pipeline, new_params)
                
                X_train = self._X_train.copy()
                y_train = self._y_train.copy()
                X_train = X_train.reset_index(drop=True)
                y_train = y_train.reset_index(drop=True)
                
                self.best_pipeline.fit(X_train, y_train)
```

### Класс для создания пайплайна для регрессии

```python
'''
    Создаём класс библиотеки optuna в котором прописываем пайплайн, 
    вместе с переменными параметрами, которые будем подбирать для регрессии
'''
class ObjectiveRegression(Objective):
    def __init__(self, X_train, y_train, cv_num):
        super().__init__(X_train, y_train, cv_num)
        
        self.ms = MetricSmape()
        self.kf = KFold(n_splits=self.cv_num, shuffle=True, random_state=RANDOM_STATE)
        
        # Создаём словарь для кодирования категорий с помощью OrdinalEncoder
        self.category_order = {
            'dept': ['hr', 'marketing', 'purchasing', 'technology', 'sales'],
            'workload': ['low', 'medium', 'high'],
            'level': ['junior', 'middle', 'sinior'],
            'last_year_promo': ['no', 'yes'],
            'last_year_violations': ['no', 'yes'],
        }
        
        
    # Задаём тренировочную выборку
    def set_model(self, model_name):
        if model_name == 'LinearRegression':
            self.model_name = 'LinearRegression'
        elif model_name == 'DecisionTreeRegressor':
            self.model_name = 'DecisionTreeRegressor'
        elif model_name == 'RandomForestRegressor':
            self.model_name = 'RandomForestRegressor'
        elif model_name == 'KNNRegressor':
            self.model_name = 'KNNRegressor'
        elif model_name == 'Ridge':
            self.model_name = 'Ridge'
        elif model_name == 'CatBoostRegressor':
            self.model_name = 'CatBoostRegressor'
        else:
            print(f'Модель не была установлена! Модели {model_name} нет в списке.')
    
    
    # Задаём метрику для поиска лучшей модели
    def set_score(self, score):
        self.score_name = score
        if self.score_name == 'MAE':
            self.score = make_scorer(mean_absolute_error, greater_is_better=False)
        elif self.score_name == 'SMAPE':
            self.score = self.ms.smape
    
    
    # В задаче регрессии не применяется семплирование, но поскольку у нас общий пайплайн, ставим passthrough
    def create_sampler_params(self, trial):
        self._sampler = Passthrough()
    
    
    # Метод для создания моделей и вариантов параметров модели (для подбора)
    def create_model_with_model_params(self, trial):        
        if self.model_name == 'KNNRegressor':
            self._knn_params = {
                'n_neighbors': trial.suggest_int('n_neighbors', 5, 100),
                'weights': trial.suggest_categorical('weights', ['uniform', 'distance']),
                'algorithm': trial.suggest_categorical('algorithm', ['auto', 
                                                                     'ball_tree', 
                                                                     'kd_tree', 
                                                                     'brute'])
            }
            self._model = KNeighborsRegressor(**self._knn_params)
        
        elif self.model_name == 'LinearRegression':
            self.linear_regression_params = {
                'fit_intercept': trial.suggest_categorical('fit_intercept', [True, False])
            }
            self._model = LinearRegression(**self.linear_regression_params)
        
        elif self.model_name == 'DecisionTreeRegressor':
            self._decisiontree_params ={
                'random_state': RANDOM_STATE,
                'max_depth': trial.suggest_int('max_depth', 2, 20),
                'max_features': trial.suggest_int('max_features', 2, 12),
                'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
                'criterion': trial.suggest_categorical('criterion', ['squared_error', 
                                                                     'friedman_mse', 
                                                                     'absolute_error', 
                                                                     'poisson'])
            }
            self._model = DecisionTreeRegressor(**self._decisiontree_params)
        
        elif self.model_name == 'RandomForestRegressor':
            self._random_forest_params ={
                'random_state': RANDOM_STATE,
                'max_depth': trial.suggest_int('max_depth', 2, 20),
                'max_features': trial.suggest_int('max_features', 2, 12),
                'criterion': trial.suggest_categorical('criterion', ['squared_error', 
                                                                     'friedman_mse', 
                                                                     'poisson'])
            }
            self._model = RandomForestRegressor(**self._random_forest_params)
            
        elif self.model_name == 'Ridge':
            self._ridge_params ={
                'random_state': RANDOM_STATE,
                'alpha': trial.suggest_float('alpha', 1e-2, 1e2, log=True),
                'fit_intercept': trial.suggest_categorical('fit_intercept', [True, False]),
                'solver': trial.suggest_categorical('solver', ['auto', 'svd', 'cholesky', 'lsqr', 'sag', 'saga'])
            }
            self._model = Ridge(**self._ridge_params)
            
        elif self.model_name == 'CatBoostRegressor':
            self._catboost_params ={
                'random_seed': RANDOM_STATE,
                'iterations': 10000,
                'verbose': False,
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
                'depth': trial.suggest_int('depth', 2, 10),
                'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-2, 1e2, log=True),
                'random_strength': trial.suggest_float('random_strength', 0, 1),
                'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
                'loss_function': trial.suggest_categorical('loss_function', ['MAE', 'RMSE', 'MAPE'])
            }
            self._model = CatBoostRegressor(**self._catboost_params)
```

### Класс для создания пайплайна для классификации

```python
'''
    Создаём класс библиотеки optuna в котором прописываем пайплайн, 
    вместе с переменными параметрами, которые будем подбирать для классификации
'''
class ObjectiveClassifier(Objective):
    def __init__(self, X_train, y_train, cv_num):
        super().__init__(X_train, y_train, cv_num)
        
        self.kf = StratifiedKFold(n_splits=cv_num, shuffle=True, random_state=RANDOM_STATE)
        
        # Создаём словарь для кодирования категорий с помощью OrdinalEncoder
        self.category_order = {
            'workload': ['low', 'medium', 'high'],
            'level': ['junior', 'middle', 'sinior'],
            'last_year_promo': ['no', 'yes'],
            'last_year_violations': ['no', 'yes'],
        }
    
    
    # Задаём тренировочную выборку
    def set_model(self, model_name):
        if model_name == 'SVC':
            self.model_name = 'SVC'
        elif model_name == 'KNNClassifier':
            self.model_name = 'KNNClassifier'
        elif model_name == 'LogisticRegression':
            self.model_name = 'LogisticRegression'
        elif model_name == 'DecisionTreeClassifier':
            self.model_name = 'DecisionTreeClassifier'
        elif model_name == 'RandomForestClassifier':
            self.model_name = 'RandomForestClassifier'
        elif model_name == 'CatBoostClassifier':
            self.model_name = 'CatBoostClassifier'
        else:
            print('Модель не была установлена! Такой модели нет в списке')
    
    # Задаём метрику для поиска лучшей модели
    def set_score(self, score):
        self.score_name = score
        if self.score_name == 'roc_auc':
            self.score = make_scorer(roc_auc_score, greater_is_better=True, needs_proba=True)
        elif self.score_name == 'accuracy':
            self.score = make_scorer(accuracy_score, greater_is_better=True)
        elif self.score_name == 'f1':
            self.score = make_scorer(f1_score, greater_is_better=True)
    
    
    # Метод для выбора способа сэмплирования (для подбора параметров)
    def create_sampler_params(self, trial):
        self._sampler_param = {
            'sampler' : trial.suggest_categorical('sampler', ['RandomUnderSampler', 
                                                              'SMOTE', 
                                                              'ADASYN', 
                                                              'SMOTETomek', 
                                                              'passthrough'])
        }
        
        if self._sampler_param['sampler'] == 'RandomUnderSampler':
            self._sampler = RandomUnderSampler(random_state=RANDOM_STATE)
        elif self._sampler_param['sampler'] == 'SMOTE':
            self._sampler = SMOTE(random_state=RANDOM_STATE)
        elif self._sampler_param['sampler'] == 'ADASYN':
            self._sampler = ADASYN(random_state=RANDOM_STATE)
        elif self._sampler_param['sampler'] == 'SMOTETomek':
            self._sampler = SMOTETomek(random_state=RANDOM_STATE)
        else:
            self._sampler = Passthrough()
        

    # Метод для создания модели и параметров модели (для подбора параметров)
    def create_model_with_model_params(self, trial):
        if self.model_name == 'SVC':
            self._svc_params ={
                'random_state': RANDOM_STATE,
                'probability': True,
                'C': trial.suggest_int('C', 1, 4),
                'kernel': trial.suggest_categorical('kernel', ['poly', 'rbf', 'sigmoid'])
            }
            self._model = SVC(**self._svc_params)
        
        elif self.model_name == 'KNNClassifier':
            self._knn_params = {
                'n_neighbors': trial.suggest_int('n_neighbors', 2, 100),
                'weights': trial.suggest_categorical('weights', ['uniform', 'distance']),
                'algorithm': trial.suggest_categorical('algorithm', ['auto', 
                                                                     'ball_tree', 
                                                                     'kd_tree', 
                                                                     'brute'])
            }
            self._model = KNeighborsClassifier(**self._knn_params)
        
        elif self.model_name == 'LogisticRegression':
            self._logistic_regression_params ={
                'random_state': RANDOM_STATE,
                'C': trial.suggest_int('C', 2, 20)
            }
            solver_penalty_comb = trial.suggest_categorical('solver_and_penalty', 
                                                            ['lbfgs + l2', 'lbfgs + None', 
                                                             'liblinear + l1', 'liblinear + l2',
                                                             'sag + l2', 'sag + None',
                                                             'saga + l1', 'saga + l2', 'saga + None'])
            if solver_penalty_comb == 'lbfgs + l2':
                self._logistic_regression_params['solver'] = 'lbfgs'
                self._logistic_regression_params['penalty'] = 'l2'
            elif solver_penalty_comb == 'lbfgs + None':
                self._logistic_regression_params['solver'] = 'lbfgs'
                self._logistic_regression_params['penalty'] = None
            if solver_penalty_comb == 'liblinear + l1':
                self._logistic_regression_params['solver'] = 'liblinear'
                self._logistic_regression_params['penalty'] = 'l1'
            elif solver_penalty_comb == 'liblinear + l2':
                self._logistic_regression_params['solver'] = 'liblinear'
                self._logistic_regression_params['penalty'] = 'l2'
            elif solver_penalty_comb == 'sag + l2':
                self._logistic_regression_params['solver'] = 'sag'
                self._logistic_regression_params['penalty'] = 'l2'
            elif solver_penalty_comb == 'sag + None':
                self._logistic_regression_params['solver'] = 'sag'
                self._logistic_regression_params['penalty'] = None
            elif solver_penalty_comb == 'saga + l1':
                self._logistic_regression_params['solver'] = 'saga'
                self._logistic_regression_params['penalty'] = 'l1'
            elif solver_penalty_comb == 'saga + l2':
                self._logistic_regression_params['solver'] = 'saga'
                self._logistic_regression_params['penalty'] = 'l2'
            elif solver_penalty_comb == 'saga + None':
                self._logistic_regression_params['solver'] = 'saga'
                self._logistic_regression_params['penalty'] = None
            
            self._model = LogisticRegression(**self._logistic_regression_params)
        
        elif self.model_name == 'DecisionTreeClassifier':
            self._decisiontree_params = {
                'random_state': RANDOM_STATE,
                'max_depth': trial.suggest_int('max_depth', 2, 20),
                'max_features': trial.suggest_int('max_features', 2, 12)
            }
            self._model = DecisionTreeClassifier(**self._decisiontree_params)
        
        elif self.model_name == 'RandomForestClassifier':
            self._random_forest_params = {
                'random_state': RANDOM_STATE,
                'max_depth': trial.suggest_int('max_depth', 2, 10),
                'max_features': trial.suggest_int('max_features', 2, 10),
                'n_estimators': trial.suggest_int('n_estimators', 100, 1000)
            }
            self._model = RandomForestClassifier(**self._random_forest_params)

        elif self.model_name == 'CatBoostClassifier':
            self._catboost_params = {
                'random_seed': RANDOM_STATE,
                'iterations': 10000,
                'verbose': False,
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
                'depth': trial.suggest_int('depth', 2, 10),
                'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-2, 1e2, log=True),
                'random_strength': trial.suggest_float('random_strength', 0, 1),
                'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
            }
            if isinstance(self._sampler, Passthrough):
                self._catboost_params['auto_class_weights'] = 'Balanced'
            
            self._model = CatBoostClassifier(**self._catboost_params)
```

<a id="Класс_для_подбора"></a>
### Класс для запуска процесса подбора лучших параметров пайплайна

```python
# Класс для запуска подбора лучших параметров пайплайна
class Study:
    def __init__(self, X_train, y_train, X_test, y_test, reg_clas_flag, cv_num):
        # Создаём словари для сохранения лучших пайплайнов и результатов
        self.best_piplines_dict = {}
        self.models_res_dict = {}
        
        # Создаём необходимиые переменные (устанавливаем дефолтные значения)
        self.reg_clas_flag = reg_clas_flag
        self.score = None
        self._cv_num = cv_num
        self._trials_num=5
        self._show_trials_num = 1
        
        # Создаём переменные для тренировочной и тестовой выборок
        self._X_train = X_train
        self._y_train = y_train
        self._X_test = X_test
        self._y_test = y_test
        
        # Создаём объект класса с настроемым для поиска параметров пайплайном
        if self.reg_clas_flag == 'Regression':
            self._objective = ObjectiveRegression(self._X_train, self._y_train, self._cv_num)
        elif self.reg_clas_flag == 'Classifier':
            self._objective = ObjectiveClassifier(self._X_train, self._y_train, self._cv_num)
        else:
            print('Ошибка: reg_clas_flag не был установлен')
        
        # Создаём объект с метриками
        self.ms = MetricSmape()
    
    
    # Задаём список моделей
    def set_models_list(self, models_list):
        self.models_list = models_list
        
    # Задаём метрику
    def set_model_score(self, score):
        self.score = score

    # Задаём количество поисков для одной модели
    def set_trials_num(self, trials_num):
        self._trials_num = trials_num

    # Задаём количество выводимых на экран результатов поиска
    def set_show_trials_num(self, show_trials_num):
        self._show_trials_num = show_trials_num
        
    # Задаём тренировочную и тестовую выборки
    def set_dataframe(self, X_train, y_train, X_test, y_test):
        self._X_train = X_train
        self._y_train = y_train
        self._X_test = X_test
        self._y_test = y_test
    
    
    # Функция для запуска процесса подбора гиперпараметров для пайплайна
    def run(self):
        # Создаём датафрейм, для вывода на экрран результата всех моделей
        self._study_info = pd.DataFrame(columns=['model', f'{self.score}_cv', f'{self.score}_test'])
        self._study_info.columns.name = 'models_best_results'
        
        # Добавляем в объект класса Objective тренировочную выборку и указываем метрику
        self._objective.set_train_and_test_data(self._X_train, self._y_train)
        self._objective.set_score(self.score)
        
        # Проходим цыклом по всем моделям
        for model_name in self.models_list:
            self._objective.set_model(model_name)
            
            # Создаём объект класса potuna для подбора параметров
            study = optuna.create_study(direction='maximize', 
                                        study_name='test_models', 
                                        sampler=TPESampler(seed=RANDOM_STATE))
            
            # Запускем процесс подбора параметров
            study.optimize(self._objective, 
                           n_trials=self._trials_num,
                           callbacks=[self._objective.callback], 
                           n_jobs=1, 
                           show_progress_bar=True)
            
            # Создаём датафрейм с результатами триалов, сохраняем в словарь, выводим на экран
            model_res = self.create_model_res_dataframe(model_name, study.trials_dataframe())
            # Убираем "user_attrs_" из назваания добавленного для catboost параматра с количеством итераций
            if model_name in ['CatBoostRegressor', 'CatBoostClassifier']:
                model_res.rename(columns=lambda x: x.replace('user_attrs_', ''), inplace=True)
            model_res = model_res.reset_index(drop=True)
            
            # Сохраняем датафрейм с результатами триалов в словарь
            self.models_res_dict[model_name] = model_res
            
            # Выводим датафрейм с результатами триалов на экран (при необходимости)
            if self._show_trials_num:
                model_res_show = model_res.copy()
                model_res_show.index = range(1, len(model_res_show) + 1)
                display(model_res_show.head(self._show_trials_num))

            # Обучаем лучший пайплайн
            self._objective.best_pipeline.fit(self._X_train.reset_index(drop=True), 
                                              self._y_train.reset_index(drop=True))
            
            # Добавляем лучший пайплайн, уже обученный, в словарь
            self.best_piplines_dict[model_name] = self._objective.best_pipeline
            
            # Получаем предсказания на тестовой выборке
            self._y_pred = self._objective.best_pipeline.predict(self._X_test)
            
            # Для классификации предсказываем вероятность
            if self.reg_clas_flag == 'Classifier':
                self._y_proba = self._objective.best_pipeline.predict_proba(self._X_test)[:,1]

            # Записываем лучшие метрики регрессии в датафрейм
            if self.score == 'MAE':
                self._study_info.loc[-1] = [model_name, model_res.loc[0, self.score], 
                                            mean_absolute_error(self._y_test, self._y_pred)]
            elif self.score == 'SMAPE':
                self._study_info.loc[-1] = [model_name, model_res.loc[0, self.score], 
                                            abs(self.ms.smape(self._objective.best_pipeline, 
                                                              self._X_test, self._y_test))]
            
            # Записываем лучшие метрики классификации в датафрейм
            if self.score == 'roc_auc':
                self._study_info.loc[-1] = [model_name, model_res.loc[0, self.score], 
                                            roc_auc_score(self._y_test, self._y_proba)]
            elif self.score == 'accuracy':
                self._study_info.loc[-1] = [model_name, model_res.loc[0, self.score], 
                                            accuracy_score(self._y_test, self._y_pred)]
            elif self.score == 'f1':
                self._study_info.loc[-1] = [model_name, model_res.loc[0, self.score], 
                                            f1_score(self._y_test, self._y_pred)]
            
            self._study_info = self._study_info.reset_index(drop=True)
        
        # Сортируем значения в датафрейме с итоговыми результатами и выводим на экран
        self._study_info = self._study_info.reset_index(drop=True)
        if self.reg_clas_flag == 'Regression':
            self._study_info = self._study_info.sort_values(by=f'{self.score}_cv')
        elif self.reg_clas_flag == 'Classifier':   
            self._study_info = self._study_info.sort_values(by=f'{self.score}_cv', ascending=False)
        print()
        self._study_info = self._study_info.reset_index(drop=True)
        study_info_show = self._study_info.copy()
        study_info_show.index = range(1, len(study_info_show) + 1)
        display(study_info_show)

        
    # Функция для преобразования датафрейма для более наглядного отображения параметров модели
    def create_model_res_dataframe(self, model_name, model_res):
        # Удаляем неинформативные колонки, меняем названия, сортируем значения
        model_res = model_res.sort_values(by='value', ascending=False)
        model_res.columns.name = model_name
        model_res = model_res.reset_index(drop=True)
        model_res['duration'] = model_res['duration'].astype(str).str.replace('0 days ', '')
        model_res = model_res.drop(columns=['number', 'datetime_start', 'datetime_complete', 'state'])
        model_res.columns = model_res.columns.str.replace('params_', '')
        
        # Расставляем колонки в нужном порядке
        cols = model_res.columns.tolist()
        rest_cols = cols.copy()
        rest_cols.remove('value')
        rest_cols.remove('duration')
        rest_cols.remove('drop_workload_flag')
        rest_cols.remove('drop_level_flag')
        # Для деревьев переставляем также и 'cat_encoding'
        if 'cat_encoding' in cols:
            rest_cols.remove('cat_encoding')
        rest_cols.remove('num_encoding')
        if 'cat_encoding' in cols:
            new_cols = ['value', 
                        'duration', 
                        'drop_workload_flag', 
                        'drop_level_flag', 
                        'num_encoding', 
                        'cat_encoding'] + rest_cols
        else:
            new_cols = ['value', 
                        'duration', 
                        'drop_workload_flag',
                        'drop_level_flag', 
                        'num_encoding'] + rest_cols
        
        model_res = model_res[new_cols]
        
        # Убираем у метрики знак "-" для регрессии
        if self.reg_clas_flag == 'Regression':
            model_res['value'] = abs(model_res['value'])
        
        # Меняем название колонки с результатом на нашу метрику
        model_res = model_res.rename(columns={'value': self.score})
        
        return model_res
```

### Класс для интерпретации модели

```python
# Класс для анализа важности признаков с помошью shap
class ShapShow:
    def __init__(self, pipeline, X_test, y_test):
        self.pipeline = pipeline
        self.X_test = X_test
        self.y_test = y_test
        self.feature_names = self.X_test.columns.tolist()
        self.cat_feature_names = self.X_test.select_dtypes(include=['object']).columns.tolist()
        self.num_feature_names = self.X_test.select_dtypes(exclude=['object']).columns.tolist()
        self.create_index_dict()
        self.create_shap_values()
    
    
    # Создаём датафрейм с категориями ошибок и id пользователей
    def create_index_dict(self):
        self.test_res = self.y_test.copy()
        self.test_res = self.test_res.reset_index().reset_index()
        self.test_res = self.test_res.rename(columns={'quit': 'y_true'})
        self._y_pred = self.pipeline.predict(self.X_test)
        self._y_pred_proba = self.pipeline.predict_proba(self.X_test)[:,1]
        self.test_res['y_pred'] = self._y_pred.tolist()
        self.test_res['y_pred_proba'] = self._y_pred_proba.tolist()
        self.test_res['mistake'] = self.test_res.apply(lambda x: self.set_mistake_cat(x.y_true, x.y_pred), axis=1)
    
    
    # Вспомогательнвая функция, что бы промаркировать ошибки
    def set_mistake_cat(self, real, pred):
        if (pred == 0) and (real == 0):
            return 'TN'
        elif (pred == 1) and (real == 1):
            return 'TP'
        elif (pred == 0) and (real == 1):
            return 'FN'
        elif (pred == 1) and (real == 0):
            return 'FP'
    
    
    # Метод для создания shap-значений
    def create_shap_values(self):
        # Подготавливаем тестовые данные таким же образом, как в лучшем триале модели
        self.X_test_preprocessed = self.pipeline.named_steps['preprocessor'].transform(self.X_test)

        # Сохраняем в переменную название использовавшегося кодировщика
        cat_encoding = self.pipeline['preprocessor'].transformers_[0][1].steps[1][1]
        
        # Если данные были закодированы OneHotEncoder
        if isinstance(cat_encoding, OneHotEncoder):
            cat_cols = self.pipeline['preprocessor'].transformers_[0][1][1].get_feature_names_out().tolist()
            num_cols = self.pipeline['preprocessor'].transformers_[1][2]
            cols = cat_cols + num_cols
            self.X_test_preprocessed = pd.DataFrame(self.X_test_preprocessed, columns=cols)
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('x0', 'level')
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('x1', 'workload')
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('x2', 'last_year_promo')
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('x3', 'last_year_violations')
            
        
        # Если данные были закодированы EmbeddingEncoder
        elif isinstance(cat_encoding, EmbeddingEncoder):
            # Получаем названия закодированных колонок в нужном порядке
            cat_cols = self.pipeline['preprocessor'].transformers_[0][1][1].get_feature_names_out().tolist()
            num_cols = self.pipeline['preprocessor'].transformers_[1][2]
            cols = cat_cols + num_cols
            # # Создаём датафрейм с закодированными признаками
            self.X_test_preprocessed = pd.DataFrame(self.X_test_preprocessed, columns=cols)
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('col_0_dim', 'level')
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('col_1_dim', 'workload')
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('col_2_dim', 'last_year_promo')
            self.X_test_preprocessed.columns = self.X_test_preprocessed.columns.str.replace('col_3_dim', 'last_year_violations')
            self.X_test_preprocessed = pd.DataFrame(self.X_test_preprocessed)
        
        # Если данные были закодированы другими колировщиками
        else:
            # Получаем названия закодированных колонок в нужном порядке
            cat_cols = self.pipeline['preprocessor'].transformers_[0][2] 
            num_cols = self.pipeline['preprocessor'].transformers_[1][2]
            cols = cat_cols + num_cols
            # Создаём датафрейм с закодированными признаками
            self.X_test_preprocessed = pd.DataFrame(self.X_test_preprocessed, columns=cols)
        
        # Создание объяснителя SHAP на основе модели
        self.explainer = shap.TreeExplainer(self.pipeline.named_steps['model'])
        # Вычисление значений SHAP для предварительно обработанного тестового набора данных
        self.shap_values = self.explainer.shap_values(self.X_test_preprocessed)
        
        # Создаём новык shap_values и shap_explanation, для кодировщиков, 
        # которые увеличивают количество категориальных признаков
        if (isinstance(cat_encoding, EmbeddingEncoder) | isinstance(cat_encoding, OneHotEncoder)):
            # Восстанавливаем исходное количество категориальных признаков
            self.shap_values = self.aggregate_shap_values(self.shap_values, 
                                                          self.X_test_preprocessed.columns.tolist(), 
                                                          self.feature_names, 
                                                          self.num_feature_names)
        
            # Создание объекта SHAP Explanation, с оббъединёнными значениями shap_values
            self.shap_exp = shap.Explanation(values=self.shap_values, 
                                             base_values=self.explainer.expected_value,
                                             feature_names=self.X_test.columns.tolist(),
                                             data=self.X_test.values)
            
        else:
            # Создание объекта SHAP Explanation, с оригинальными значениями shap_values
            self.shap_exp = shap.Explanation(values=self.shap_values, 
                                             base_values=self.explainer.expected_value,
                                             feature_names=self.X_test_preprocessed.columns.tolist(),
                                             data=self.X_test_preprocessed.values)
        
        
    # Метод для бъединение значений SHAP для категориальных признаков
    def aggregate_shap_values(self, shap_values, encoded_feature_names, original_feature_names, numerical_features):
        # Создаём двумерный массив нулей, размер: количество строк, реальное количество категориальных признаков
        aggregated_shap_values = np.zeros((shap_values.shape[0], len(original_feature_names)))
        
        # Проходим циклом по всем признакам
        for i, original_feature in enumerate(original_feature_names):
            # Для количественных признаков просто копируем значения SHAP
            if original_feature in numerical_features:
                mask = np.array(encoded_feature_names) == original_feature
                aggregated_shap_values[:, i] = shap_values[:, mask].sum(axis=1)
            # Для категориальных признаков суммируем значения SHAP по закодированным признакам
            else:
                mask = np.array([encoded_feature.startswith(original_feature) for encoded_feature in encoded_feature_names])
                # Проверка, есть ли соответствие
                if mask.sum() > 0:
                    aggregated_shap_values[:, i] = shap_values[:, mask].sum(axis=1)
                else:
                    raise ValueError(f"No encoded features found for original feature '{original_feature}'")
                    
        return aggregated_shap_values
    
    
    # Метод для вывода графика barplot с важностью признаков
    def show_barplot(self):
        shap.summary_plot(self.shap_exp, plot_type='bar', show=False)
        plt.gcf().set_size_inches(5, 3)
        plt.gca().tick_params(labelsize=9)
        plt.gca().set_xlabel('mean(|SHAP value|)', fontsize=9)
        plt.grid(True)
        plt.show()
    
    
    # Метод для вывода графика beeswarm c общей зависимостьи предсказаний от признаков
    def show_summary_plot(self):
        shap.summary_plot(self.shap_exp, plot_type='dot', show=False)
        plt.gcf().set_size_inches(5, 3)
        plt.gca().tick_params(labelsize=9)
        plt.gca().set_xlabel('SHAP value (impact on model output)', fontsize=9)
        plt.grid(True)
        plt.show()
    
    
    # Метод для вывода графика waterfall для конкретного пользователя
    def show_waterfall(self, index):
        display(self.test_res.loc[[index]])

        # Получаем начальные данные для графика
        original_shap_values = self.shap_exp[index].values
        base_rate = self.y_test.mean().values[0]
        predicted_prob = self.test_res.loc[index, 'y_pred_proba']

        # Вычисляем коэффициент масштабирования
        total_effect = original_shap_values.sum()
        scaling_factor = (predicted_prob - base_rate) / total_effect

        # Преобразуем значения
        converted_shap_values = original_shap_values * scaling_factor

        # Создаём массив для графика водопада
        waterfall_values = np.concatenate([[base_rate], converted_shap_values, [predicted_prob]])

        # Создаём объект SHAP Explanation для графика
        new_shap_exp = shap.Explanation(
            values=converted_shap_values,
            base_values=base_rate,
            feature_names=self.shap_exp.feature_names,
            data=self.shap_exp.data[index]
        )

        # Строим график водопада
        shap.waterfall_plot(new_shap_exp, show=False)
        plt.gcf().set_size_inches(5, 3)
        plt.gca().tick_params(labelsize=9)
        for text in plt.gcf().findobj(match=plt.Text):
            text.set_fontsize(9)
        plt.show()
    
    
    # Метод для вывода на экран пользователей с определённой категорией ошибки модели прогноза
    def show_mistake_samples(self, mistake, num_samples=5):
        if mistake == 'FN':
            df = self.test_res[self.test_res['mistake'] == 'FN']
            df = df.sort_values(by='y_pred_proba').reset_index(drop=True)
            print(f'Всего ошибок типа {mistake}: {len(df)}')
            display(df.head(num_samples))
        elif mistake == 'FP':
            df = self.test_res[self.test_res['mistake'] == 'FP']
            print(f'Всего ошибок типа {mistake}: {len(df)}')
            df = df.sort_values(by='y_pred_proba', ascending=False).reset_index(drop=True)
            display(df.head(num_samples))
        elif mistake == 'TN':
            df = self.test_res[self.test_res['mistake'] == 'TN'].reset_index(drop=True)
            print(f'Всего ситуаций типа {mistake}: {len(df)}')
            display(df.head(num_samples))
        elif mistake == 'TP':
            df = self.test_res[self.test_res['mistake'] == 'TP'].reset_index(drop=True)
            print(f'Всего ситуаций типа {mistake}: {len(df)}')
            display(df.head(num_samples))
    
   
    # Метод для построения матрицы ошибок
    def show_confusion_matrix(self): 
        # Строим матрицу ошибок
        cm = confusion_matrix(self.y_test, self._y_pred)
        plt.figure(figsize=(4, 3))
        plt.rc('font', size=10) 
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
        plt.title('Матрица ошибок')
        plt.ylabel('True label')
        plt.xlabel('Predicted')
        plt.show()
        # Выводим метрики
        acc = accuracy_score(self.y_test, self._y_pred)
        precision = precision_score(self.y_test, self._y_pred)
        recall = recall_score(self.y_test, self._y_pred)
        print(f'accuracy = {acc:.4}')
        print(f'precision = {precision:.4}')
        print(f'recall = {recall:.4}')
    
    # Метод для построения графика roc-auc кривой
    def show_roc_curve(self): 
        # Считаем кривую
        fpr, tpr, treshold = roc_curve(self.y_test, self._y_pred_proba)
        roc_auc = auc(fpr, tpr)
        # Строим кривую
        plt.figure(figsize=(5, 3))
        plt.plot(fpr, tpr, color='darkorange',
                 label='ROC кривая (area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC-кривая')
        plt.legend(loc="lower right")
        plt.grid()
        plt.show()
    
    
    # Метод для построения графика зависимости количества ошибок от порога
    def show_mistakes_threshold_plot(self):
        # Проходим циклом по пороговым значениям и считаем кол-во ошибок
        data_thresholds = []
        for threshold in np.arange(0, 1, 0.005):
            precisions = len(self.test_res[(self.test_res['y_pred_proba'] > threshold) & 
                                           (self.test_res['y_true'] == 0)])
            recalls = len(self.test_res[(self.test_res['y_pred_proba'] <= threshold) & 
                                        (self.test_res['y_true'] == 1)])
            data_thresholds.append([precisions, recalls, round(threshold, 3)])
        # Сохраняем количества ошибок и соответствующие им пороговые значения в виде датафрейма
        data_thresholds = pd.DataFrame(data_thresholds, columns=['precisions', 'recalls', 'threshold'])
        # Строим график
        fig, ax = plt.subplots(figsize=(8, 4))
        sns.lineplot(data=data_thresholds, x='threshold', y='precisions', label='False Positive', ax=ax)
        sns.lineplot(data=data_thresholds, x='threshold', y='recalls', label='False Negative', ax=ax)
        ax.set(title='Зависимость количества ошибок от порога', 
               ylabel='количество ошибок', 
               xlabel='пороговое значение')
        ax.legend()
        plt.grid()
        plt.show()
```

#### Выбор метрики для оценки модели

Для оценки модели была выбрана ментрика SMAPE, которая рассчитывается по формуле:

$$ 
SMAPE = \frac{100}{n} \sum_{i=1}^n \frac{|y_i - \hat{y}_i|}{\frac{(|y_i| + |\hat{y}_i|)}{2}} ,
$$

где:
- $ y_i $ - фактическое значение целевого признака
- $ \hat{y}_i $ - предсказанне значение целевого признака
- $ n $ - количество объектов в выборке

Критерий успеха: $ SMAPE \le 15 $ на тестовой выборке.

Данная метрика создаётся в отдельном классе Scores.

Поскольку, поиск работает достаточно долго, для экономии времени при работе над второй частью проекта, создадим модель вручную.

```python
# Создаём эеземпляр класса с метрикой
ms = MetricSmape()

# Создаём пайплайн по заданным параметрам
def create_regressor_model(encoder):
    cat_pipe = ImbPipeline(
        [
            ('simple_imputer_before_cat', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
            ('cat_encoder', encoder),
            ('simple_imputer_after_cat', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
        ]
    )

    num_pipe = ImbPipeline(
        [
            ('simple_imputer_before_num', SimpleImputer(missing_values=np.nan, strategy='median')),
            ('num_encoder', MinMaxScaler())
        ]
    )

    preprocessor = ColumnTransformer(
        [
            ('cat', cat_pipe, make_column_selector(dtype_include=object)),
            ('num', num_pipe, make_column_selector(dtype_include=np.number))
        ], 
        remainder='passthrough'
    )

    pipe_regressor = ImbPipeline(
        [
            ('preprocessor', preprocessor),
            ('model', CatBoostRegressor(random_seed=RANDOM_STATE, 
                                        iterations=1139, 
                                        verbose=False,
                                        bagging_temperature=0.264570, 
                                        depth=7, 
                                        l2_leaf_reg=0.245743, 
                                        learning_rate=0.030131, 
                                        random_strength=0.323693))
        ]
    )
    
    return pipe_regressor

best_pipe_regressor = create_regressor_model(EmbeddingEncoder(embedding_dim=26, 
                                                              handle_unknown='ignore', 
                                                              random_state=RANDOM_STATE))
best_pipe_regressor_ohe = create_regressor_model(OneHotEncoder(drop='first', 
                                                               handle_unknown='ignore'))

# Обучаем и повторно тестируем лучшую модель
best_pipe_regressor.fit(X_train, y_train)
y_pred = best_pipe_regressor.predict(X_test)
smape_test = abs(ms.smape(best_pipe_regressor, X_test, y_test))
print(f'Модель Catboost, метрика smape на тестовой выборке: {smape_test}')

# Обучаем и повторно тестируем лучшую модель
best_pipe_regressor_ohe.fit(X_train, y_train)
y_pred = best_pipe_regressor.predict(X_test)
smape_test = abs(ms.smape(best_pipe_regressor_ohe, X_test, y_test))
print(f'Модель Catboost c OHE-кодировщиком, метрика smape на тестовой выборке: {smape_test}')
```

Поскольку, поиск работает достаточно долго, для экономии времени при работе над заключительной частью проекта, создадим модель вручную. Для более наглядной интерпретации данных создадим ещё одну точно такую же модель, но с использованием OrdinalEncoder. 

```python
# Определяем категории для OrdinalEncoder
category_order = {
    'workload': ['low', 'medium', 'high'],
    'level': ['junior', 'middle', 'sinior'],
    'last_year_promo': ['no', 'yes'],
    'last_year_violations': ['no', 'yes'],
}
cat_features = X_train_quit.select_dtypes(include=['object']).columns.tolist()
categories = [category_order[col] for col in cat_features]

def create_classifier_model(encoder):
    # Создаём пайплайн по заданным параметрам
    cat_pipe = ImbPipeline(
        [
            ('simple_imputer_before_encoder', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
            ('encoder', encoder),
            ('simple_imputer_after_encoder', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
        ]
    )

    num_pipe = ImbPipeline(
        [
            ('simple_imputer_before_scaler', SimpleImputer(missing_values=np.nan, strategy='median')),
            ('scaler', StandardScaler())
        ]
    )

    preprocessor = ColumnTransformer(
        [
            ('cat', cat_pipe, make_column_selector(dtype_include=object)),
            ('num', num_pipe, make_column_selector(dtype_include=np.number))
        ], 
        remainder=Passthrough()
    )

    pipe_classifier = ImbPipeline(
        [
            ('preprocessor', preprocessor),
            ('sampler', RandomUnderSampler(random_state=RANDOM_STATE)),
            ('model', CatBoostClassifier(random_seed=RANDOM_STATE, 
                                         iterations=198, 
                                         verbose=False,
                                         bagging_temperature=0.489742, 
                                         depth=2, 
                                         l2_leaf_reg=0.746038, 
                                         learning_rate=0.094673, 
                                         random_strength=0.677138))
        ]
    )
    
    return pipe_classifier

# Обучаем пайплайны с разными кодировщиками
best_pipe_classifier = create_classifier_model(EmbeddingEncoder(embedding_dim=37, 
                                                                handle_unknown='ignore', 
                                                                random_state=RANDOM_STATE))
best_pipe_classifier_ord = create_classifier_model(OrdinalEncoder(categories=categories, 
                                                                  handle_unknown='use_encoded_value', 
                                                                  unknown_value=-1))

# Обучаем и повторно тестируем лучшую модель
best_pipe_classifier.fit(X_train_quit, y_train_quit)
y_pred_proba = best_pipe_classifier.predict_proba(X_test_quit)[:,1]
auc_test = roc_auc_score(y_test_quit, y_pred_proba)
print(f'Модель catboost, метрика roc-auc на тестовой выборке: {auc_test}')

# Обучаем и повторно тестируем лучшую модель, но с OrdinalEncoder
best_pipe_classifier_ord.fit(X_train_quit, y_train_quit)
y_pred_proba_ord = best_pipe_classifier_ord.predict_proba(X_test_quit)[:,1]
auc_test_ord = roc_auc_score(y_test_quit, y_pred_proba_ord)
print(f'Модель catboost_ord, метрика roc-auc на тестовой выборке: {auc_test_ord}')
```
